```python
import os
import cv2
import pandas as pd
import random
import matplotlib.pyplot as plt
```

```python
GT_DIR = "train.csv"
```

```python
def show_random_patch(path):
    df = pd.read_csv(path)
    
    patches = df.groupby(["path"])
    
    grp = random.choice(list(patches.groups.keys()))
    random_patch = patches.get_group(grp)
    random_patch = random_patch.reset_index()
    
    image = None
    for i, row in random_patch.iterrows():
        if i == 0:
            image = cv2.imread(row['path'])

        start = (row['y1'], row['x1'])
        end = (row['y2'], row['x2'])

        image = cv2.rectangle(image, start, end, (0, 0, 255), 1)

    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.show()
```

```python
show_random_patch(GT_DIR)
```
